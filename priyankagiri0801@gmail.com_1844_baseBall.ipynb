```python
from scipy.stats import zscore
z_score = np.abs(zscore(data))
z_score 
np.where(z_score>3)
df_new= data[(z_score < 3).all(axis=1)]
df_new.shape
```

# SEPERATING COLUMNS AND TARGET FROM DATAFRAME

```python
features = df_new.drop('W',axis=1)
target= df_new['W']
```

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
scalar_data = scalar.fit_transform(features)
```

```python
features_train,features_test,target_train,target_test = train_test_split(scalar_data,target,test_size=0.20, random_state=42)
lr.fit(features_train,target_train)
pred_train = lr.predict(features_train)
pred_test = lr.predict(features_test)
print("accuracy of training data")
print(r2_score(target_train,pred_train))
print("accuracy of testing data")
print(r2_score(target_test,pred_test))  
```

# cross validation of model

```python
train_accuracy = r2_score(target_train,pred_train)
test_accuracy = r2_score(target_test,pred_test)
from sklearn.model_selection import cross_val_score
cv_score = cross_val_score(lr,features,target,cv=5)
cv_mean = cv_score.mean()
print("cv_mean",cv_mean)
print("train_accuracy",train_accuracy)
print("test_accuracy",train_accuracy)
```

# ENSEMBLE TECHNIQUE

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
```

```python
parameters = {'criterion':['mse','mae'],
             'max_features':['auto','sqrt','log2']}
```

```python
rf = RandomForestRegressor()
clf = GridSearchCV(rf,parameters)
clf.fit(features_train,target_train)
print(clf.best_params_)
```

```python
rf = RandomForestRegressor(criterion = 'mse', max_features='auto')
rf.fit(features_train,target_train)
rf.score(features_train,target_train)
pred_dec = rf.predict(features_test)
```

```python
rfs = r2_score(target_test,pred_dec)
print(rfs*100)
```
